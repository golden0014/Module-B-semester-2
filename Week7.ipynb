```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
data = pd.DataFrame({
    'Design Concern': ['Engine power is less than expected', 'Tires make excessive noise while driving', 'Engine makes abnormal/excessive noise', 'Seat material concerns', 'Excessive wind noise', 'Hesitation or delay when shifting', 'Bluetooth system has poor sound quality', 'Steering system/wheel has too much play', 'Bluetooth system is difficult to use', 'Front seat audio/entertainment/navigation controls'],
    'Concerns per 1,000': [12.9, 12.3, 11.6, 11.6, 11.0, 10.3, 10.0, 8.8, 8.6, 8.2]
})

ax = sns.barplot(data = data, y = 'Design Concern', x = 'Concerns per 1,000', linewidth = 3)
ax.set_title('Top 10 Design Concerns', loc = 'left', x = -0.75, y = 1, fontsize = 14, pad = 2)
ax.xaxis.set_label_position('top')
ax.set_xlabel('Concerns per 1,000', loc = 'left', fontsize = 10)
ax.tick_params(
    axis="x",
    which="major",
    bottom=False,
    labelbottom=False,
)
ax.set_ylabel('')
for i, bar in enumerate(ax.patches):
    # Color bars
    if i < 7:
        bar.set_facecolor('lightcoral')
    else:
        bar.set_facecolor('gray')
    # Add values
    value = bar.get_width()
    y = bar.get_y() + bar.get_height() / 2

    ax.text(
        value - 1,          
        y,
        f"{value:.1f}",       
        va="center",
        ha="left",
        color = 'white',
        fontsize=10
    )

ax.text(-0.75, 1.15, '7 of the Top 10 Design Concerns Have 10 or More Concerns per 1,000.', color = 'lightcoral', transform = ax.transAxes)
ax.text(-0.75, 1.09, 'Discussion: Is This an Acceptable Default Rate?', color = 'gray', transform = ax.transAxes)
```
